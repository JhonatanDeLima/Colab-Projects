```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
print(df.head(20))
```

```python
plt.figure(figsize=(12,6))
plt.bar(df['Time'], df['Pts'], color='blue')
plt.xticks(rotation=90)
plt.title('Pontos - Premier League')
plt.xlabel('Times')
plt.show()
```

```python
cores = ['green' if x >= 0 else 'red' for x in df['SG']]

plt.figure(figsize=(12,6))
plt.bar(df['Time'], df['SG'], color=cores)
plt.xticks(rotation=90)
plt.title('Saldo de Gols - Premier League')
plt.xlabel('Times')
plt.ylabel('Saldo')

min_y = min(df['SG'])
max_y = max(df['SG'])
plt.yticks(np.linspace(min_y, max_y, 10))

plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.bar(df["Time"], df["V"])
plt.xticks(rotation=90)
plt.title("Vitórias - Premier League")
plt.xlabel("Times")
plt.ylabel("Vitórias")
plt.show()

```

```python
plt.figure(figsize=(12,6))
plt.bar(df["Time"], df["GP"])
plt.xticks(rotation=90)
plt.title("Gols Pró - Premier League")
plt.xlabel("Times")
plt.ylabel("Gols Pró")
```

```python
cores = ['red' if x >= 76 else 'blue' for x in df['GC']]

plt.figure(figsize=(12,6))
plt.bar(df["Time"], df["GC"],color = cores)
plt.xticks(rotation=90)
plt.title("Gols Sofridos - Premier League")
plt.xlabel("Times")
plt.ylabel("Gols Sofridos")

plt.show()
```
