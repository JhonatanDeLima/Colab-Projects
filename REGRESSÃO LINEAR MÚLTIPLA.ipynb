```python
import pandas as pd
import numpy as np
```

```python
#Ler arquivos
df = pd.read_csv('USA_Housing.csv')
```

```python
#Exibir Colunas
print(df.columns)
```

```python
print(df.head())
```

```python
#Verificando tipo de cada varíavel
print(df.dtypes)
```

```python
#Excluindo Coluna endereço
df.drop('Address',axis=1,inplace=True)
```

```python
#Deixando preço com duas casas decimais
df['Price'] = round(df['Price'],2)
print(df['Price'])
```

CRIAR MODELO E TREINAR

```python
x = df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
        'Avg. Area Number of Bedrooms', 'Area Population']]
```

```python
y = df['Price']
```

```python
#Instalando teste e treino
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
#Bilioteca de Regressão Linea Múltipla
from sklearn.linear_model import LinearRegression
```

```python
#Criando modelo de regressão
modelopreditor = LinearRegression()
```

```python
#Treinando o modelo de regressão
modelopreditor.fit(x_train,y_train)
```

```python
#Imprimir os coeficientes
coeff = pd.DataFrame(modelopreditor.coef_,x.columns,columns=['Coeficiente'])
print(coeff)
```

```python
b0 = modelopreditor.intercept_
b = modelopreditor.coef_
```

```python
#Montar a equação
equacao = f"Price = {b0:.2f}"
termos = [f"({coef:.2f} * {col})" for col, coef in zip(x.columns, b)]
equacao += " + " + " +".join(termos)
print(equacao)
```

```python
#Avaliando modelo
#renda = 65.000
#idade média = 5
#quartos = 7, dormitórios = 4
#população = 30.000
nova_instancia = [[65000,5,7,4,30000]]
```

```python
previsao_nova = modelopreditor.predict(nova_instancia)
print(previsao_nova)
```

```python
predictions = modelopreditor.predict(x_test)
```

```python
print(predictions[:5])
print(y_test.head())
```

```python
from sklearn import metrics
```

```python
print("MAE ",metrics.mean_absolute_error(y_test, predictions))
```

```python
print("MSE: ",metrics.mean_squared_error(y_test, predictions))
```

```python
print("RMSE: ", np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
mae = metrics.mean_absolute_error(y_test, predictions)
mse = metrics.mean_squared_error(y_test, predictions)
rmse = np.sqrt(mse)
```

```python
y_mean = float(np.mean(y_test))
mae_pct = (mae / y_mean) *100
rmse_pct = (mae / y_mean) *100
print("mae: ", mae_pct)
print("RMSE: ", rmse_pct)
```
