```python
#Regressão Linear Simples
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Realizar a leitura do arquivo
df = pd.read_excel('experiencia_salario.xlsx')
```

```python
df
```

```python
#Seprar as colunas em X e Y

x = df[['Experiência']]
y = df[['Salário']]
```

```python
print(x)
print(y)
```

```python
plt.scatter(x,y, color='blue',marker='o')
plt.title("Experiência x Salário")
plt.xlabel("Anos de experiência")
plt.ylabel("Salário (R$)")
plt.show()
```

```python
import seaborn as sns

dados = pd.DataFrame(df)
```

```python
sns.scatterplot(data=dados, x = "Experiência",y="Salário",s=100,
                color="red")
plt.title("Experiência x Salário")
plt.show()
```

```python
#Regressão Linear
from sklearn.linear_model import LinearRegression
```

```python
#instânciar o objeto
linearRegression = LinearRegression()
linearRegression.fit(x,y)
```

```python
plt.scatter(x,y, color='red')
plt.plot(x,y,color='blue')
plt.title("Salário x Exp")
plt.xlabel("Anos de experiência")
plt.ylabel("Salário")
plt.show()
```

```python
print(linearRegression.coef_[0])
```

```python
print(linearRegression.intercept_)
```

```python
print('y = 583.63x + 1220.00')
```

```python
#Qual o salário de alguém 13.8 anos de experiência
print(linearRegression.predict([[13.8]]))
```

```python
#Prever múltiplas entradas
exp = pd.DataFrame([[4.7],[3.4],[23.1],[17.5]],columns=["Experiência"])
```

```python
#Fazer  a previsão dos salários
print(linearRegression.predict(exp))
```
