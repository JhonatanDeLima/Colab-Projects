```python
#Regressão Logística
#Base titanic
import pandas as pd
import numpy as np
```

```python
td = pd.read_csv('titanic_data.csv')
```

```python
td.columns
```

```python
td.describe()
```

```python
#PREPARAÇÃO DE DADOS
print(td.dtypes)
```

```python
td.head(3)
```

```python
#Apagar coluna
td.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'],axis=1,inplace=True)
```

```python
td.columns
```

```python
#Valores faltantes
td.isnull().sum()
```

```python
#MÉDIA para a idade
td['Age'] = td['Age'].fillna(td.groupby('Pclass')['Age'].transform('mean'))
```

```python
#Embarque usar MODA
embarked_mode = td['Embarked'].mode()[0]
td['Embarked']=td['Embarked'].fillna(embarked_mode)
```

```python
print(td.isnull().sum())
```

```python
td.head()
```

```python
sex = pd.get_dummies(td['Sex'], drop_first=True)
print(sex)
```

```python
embark = pd.get_dummies(td['Embarked'],drop_first=True)
print(embark)
```

```python
td.drop(['Sex', 'Embarked'], axis=1, inplace=True)
```

```python
td=pd.concat([td, sex, embark], axis=1)
```

```python
print(td.head())
```

```python
td.info()
```

```python
#Aplicar a regressão logística
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    td.drop('Survived',axis=1),td['Survived'],
    test_size=0.20, random_state=0)
print(x_train)
print(y_train)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(x_train,y_train)
```

```python
predictions = logmodel.predict(x_test)
```

```python
print(predictions)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
conf_mat = confusion_matrix(y_test, predictions)
print('Matriz de confusão')
print(conf_mat)
```

```python
print(classification_report(y_test,predictions))
```

```python
exemplo = np.array([1,31,0,0,350,1,0,1]).reshape((1,-1))
print(logmodel.predict(exemplo)[0])
```
