```python
# Importando bibliotecas e dataset
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
# Carregando o Iris e separando atributos e alvo
iris = load_iris()
#Esta é uma forma concisa de atribuir os dados.
#A variável X recebe os atributos (ou características) do dataset, que são as medidas das flores. A variável y recebe o alvo, que é a espécie de flor que o modelo deve prever.
X, y = iris.data, iris.target
```

```python
print(iris)
```

```python
# Dividindo em treino e teste
#divide os dados em conjuntos de treino e teste. O parâmetro test_size=0.3 indica que 30% dos dados serão reservados para o teste, e os 70% restantes serão usados para treinar o modelo. Isso é importante para garantir que o modelo seja avaliado com dados que ele "nunca viu" antes, o que dá uma estimativa mais realista de seu desempenho.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# Modelo 1: Naive Bayes Gaussiano
#Esse tipo de modelo é adequado para dados que seguem uma distribuição normal (gaussiana),
#como é o caso das medidas de flores no dataset Iris.
from sklearn.naive_bayes import GaussianNB
modelo1 = GaussianNB()
modelo1.fit(X_train, y_train)
#calcula a acurácia (a porcentagem de acertos) do modelo. O método .score() usa os dados de teste (X_test e y_test) para prever as espécies e compara as previsões com os valores reais. O resultado é multiplicado por 100, arredondado para duas casas decimais e formatado como uma string com o símbolo de porcentagem.
precisao1 = str(round(modelo1.score(X_test, y_test) * 100, 2)) + "%"
print("A acurácia do modelo 1 foi", precisao1)
#O método .score() é chamado a partir do objeto modelo1, que é o seu modelo de machine learning. Ele avalia o desempenho do modelo usando um conjunto de dados de teste (X_test e y_test). O resultado é um número decimal entre 0 e 1, que representa a proporção de previsões corretas.
#A função round() arredonda o valor para um número específico de casas decimais. Neste caso, o 2 indica que o valor será arredondado para duas casas decimais.
```

```python
# Modelo 2: Naive Bayes Multinomial
#Este modelo é mais indicado para dados de contagem, como a frequência de palavras em textos, e não é ideal para dados como o dataset Iris, que são valores contínuos.
from sklearn.naive_bayes import MultinomialNB
modelo2 = MultinomialNB()
modelo2.fit(X_train, y_train)
precisao2 = str(round(modelo2.score(X_test, y_test) * 100, 2)) + "%"
print("A acurácia do modelo 2 foi", precisao2)
```

```python
# Modelo 3: Naive Bayes Bernoulli
#Este modelo é projetado para dados binários (como Sim ou Não, 0 ou 1), onde cada característica é uma variável booleana.
from sklearn.naive_bayes import BernoulliNB
modelo3 = BernoulliNB()
modelo3.fit(X_train, y_train)
precisao3 = str(round(modelo3.score(X_test, y_test) * 100, 2)) + "%"
print("A acurácia do modelo 3 foi", precisao3)
```
